```python
import pandas as pd
import numpy as np
```

## Persiapan Data Frame

```python
n_rows = 40
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)
df.head()
```

## Line Plot

```python
df.head()
```

```python
df.plot(kind='line')
```

```python
df[['A', 'B']].plot(kind='line')
# kolom A dan B yang hanya di plot
```

## Bar Plot

```python
df.head()
```

```python
df.plot(kind='bar')
```

```python
df[['A', 'B']].plot(kind='bar')
```

```python
df[['A', 'B']].head().plot(kind='bar')
```

```python
df[['A', 'B']].head().plot(kind='barh')
```

## Area Plot

```python
df.head()
```

```python
df.plot(kind='area')
```

```python
df[['A', 'B']].head().plot(kind='area')
```

## Box Plot

```python
df.head()
```

```python
df.plot(kind='box')
# untuk box plot tidak merepresentasikan index tetapi di ganti menjadi A, B , C, D, dan E
```

## Histogram

```python
df.head()
```

```python
df.plot(kind='hist')
```

```python
df[['A', 'B']].head().plot(kind='hist')
```

## Kernel Density Estimation (KDE)

```python
df.head()
```

```python
df.plot(kind='kde')
```

## Scatter Plot

```python
df.head()
```

```python
#scatter plot merupakan visualisasi yang sangat cocok sekali untuk melihat korelasi antara 2 variabel/2 kolom dalam suatu data frame
df.plot(x='A', y='B', kind='scatter')
# untuk scatter plot tidak cukup hanya menyertakan parameter kind='scatter' tetapi juga menambahkan x dan y
```
