# 07 Menghapus Missing value

## Persiapan Data Frame

```python
df = pd.util.testing.makeMissingDataframe().reset_index()
df.head()
```

```python
df = df.rename(columns={'index':'Z'})
df.head()
```

```python
df_backup = df.copy(deep=True)
```

## Menghapus (drop) Setiap Kolom Yang Mengandung Missing values

```python
df = df.dropna(axis='columns')
df.head()
```

## Menghapus (drop) Setiap Baris Yang Mengandung Missing values

```python
df = df_backup.copy(deep=True)
df = df.dropna(axis='rows')
df.head()
```

## Persentase Missing Values Untuk Tiap Kolom

```python
df = df_backup.copy(deep=True)
df.isna().mean()
#fungsi df.isna untuk mengumpulkan semua missing values dari setiap kolom data frame kita
#fungsi .mean mengkalkulasi rata2 dari missing values yg dimiliki oleh setiap kolom dari data frame kita
```

## Kenapa kita harus menghitung missing values?

karena hasil persentase missing values akan digunakan untuk hal selanjutnya yaitu menghapus semua kolom yg mengandung missing kolom berdasarkan threshold tertentu

## Menghapus (drop) Setiap Kolom Yang Mengandung Missing Values Berdasarkan Threshold

```python
# fungsi dari treshold adalah memberi batas untuk setiap kolom yg ingin di buang kolomnya. disini menggunakan 0.9 / 90%

treshold = len(df) * 0.9 # len(df) fungsinya untuk mencari tau jumlah raw dari data frame kita
df = df.dropna(thresh=treshold, axis='columns')
df.head()
```

# 11 Seleksi Kolom Dan Baris Pada Data Frame Menggunakan Loc

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)
df
```

## Seleksi Kolom Dan Baris Menggunakan `Loc`

```python
df.loc[[0, 3, 4], ['B', 'E']]
# bagian pertama dari loc digunakan untuk mengatur baris
# bagian kedua dari loc untuk mengatur kolom
```

## Seleksi Baris Dengan Kondisi

```python
df.loc[df['B']>10, ['B', 'D', 'E']]
# mencari data dari kolom B yang lebih dari 10
# ['B', 'D', 'E'] bagian ini jika di mysql termasuk dibagian SELECT
# df['B']>10 bagian ini termasuk dibagian WHERE
```

## Slicing Data Frame Dengan `Loc`

```python
df.loc[0:4, 'B':'D']
# 0:4 berarti melakukan slicing dari baris ke 0 sampai ke 4
# 'B':'D' melakukan slicing mulai dari kolom B sampai kolom D
```

### Perbedaan teknik slicing pada loc dengan teknik slicing pada list :

- hal yang paling membedakan antara teknik slicing pada loc bila dibandingkan dengan teknik slicing pada list adalah jika pada loc slicingnya dibagi menjadi 2 bagian besar yaitu start index dan end index. hanya saja pada loc, baik start indek maupun end index sifatnya inclusive artinya turut disertakan. oleh karena itu jika dilihat contoh diatas, index 0 dan index 4 ikut disertakan.
- teknik slicing pada list, start index nya inclusive(turut disertakan) dan end index nya exclusive(tidak disertakan). contohnya jika ingin slicing baris 0:4, maka yang ditampilkan hanya baris 0-3

# 17 Resampling Pada Data Deret Waktu (Time Series Data)

# 33 Nilai Kumulatif Pada Pandas Data Frame

## Persiapan Data Frame

```python
d = {'pemain' : ['Budi', 'Joni', 'Iwan', 'Budi', 'Budi', 'Iwan', 'Asep', 'Joni'],
     'goal' : [2, 1, 3, 1, 1, 2, 2, 3]}
df = pd.DataFrame(d)
df
```

## Menampilkan Nilai Kumulatif (cummulative)

```python
df['goal'].cumsum().to_frame()
```

```python
df['jumlah_goal_kumulatif'] = df['goal'].cumsum()
df
# index ke 4 jumlah goal kumulatifnya ada 8. di hitung dari index 0 - index 4 kemudian di jumlah. atau seperti 2 + 1 + 3 + 1 + 1
```

```python
df['jumlah_goal_kumulatif_tiap_pemain'] = df.groupby('pemain')['goal'].cumsum()
df
```

```python
df['cummax'] = df['goal'].cummax()
df
# fugsi cummax adalah mencari nilai tertinggi. contohnya nyari nilai tertinggi dari index 0-5 maka jawabannya adalah 3
```

```python
df['cummin'] = df['goal'].cummin()
df
```

```python
df['cumprod'] = df['goal'].cumprod()
df
# fungsi cumprod adalah hasil total perkalian. contohnya mencari nilai cumprod dari index 4. maka index dari 0 - 4 di kalikan
```

# 34 Mapping Data Frame Dengan Applymap

```python
df = pd.DataFrame({'jenis_kelamin' : ['Pria', 'Wanita', 'Lelaki', 'Lelaki', 'Perempuan'],
                   'usia' : [23, 21, 24, 22,21],
                   'shift' : ['pagi', 'siang', 'Malam', 'Siang', 'pagi']})
df
```

## Mapping Pada Data Frame Dengan `applymap()`

```python
df = df.applymap(lambda x: x.lower() if type(x) == str else x)
df
# pemanfaatan method applymap dengan menggunakan lambda expression. hasilnya tulisan menjadi lower case
```

## `applymap()` Dengan Dictionary

```python
mapping = {'pria' : 'L',
           'lelaki' : 'L',
           'wanita' : 'P',
           'perempuan' : 'P',
           'pagi' : 1,
           'siang' : 2,
           'malam' : 3}

df.applymap(mapping.get)

# method applymap akan dikenakan untuk keseluruhan data baik untuk data jenis kelamin, usia, maupun shift. sedangkan dict diatas hanya menunjukkan jenis kelamin dan shift
```

```python
df[['jenis_kelamin', 'shift']] = df[['jenis_kelamin', 'shift']].applymap(mapping.get)
df
# cara untuk menggunakan applymap secara spesifik dan tidak merubah nilai data kolom usia
```

# 35 Memadukan Fungsi Agregasi Dengan Transform

## Persiapan Data Frame

```python
d = {'no_nota' : [1, 1 , 1, 2, 2, 3, 4, 5],
     'kopi' : ['latte', 'cappucino', 'espresso', 'latte', 'espresso', 'cappucino', 'latte', 'espresso'],
     'harga' : [50, 60, 80, 150, 120, 60, 100, 40]}

df = pd.DataFrame(d)
df
# index 2 dan 7 harga espresso nya berbeda. ini mungkin bisa terjadi karena data kita tidak lengkap. harusnya ada kolom quantity yg menggambarkan jumlah espresso yg dipesan
```

## Menghitung Total Harga Untuk Tiap Nomor Nota

```python
df.groupby('no_nota')['harga'].sum().to_frame()
```

```python
df['total_harga'] = df.groupby('no_nota')['harga'].transform(sum)
df
```

## Menghitung Total Omset Untuk Tiap Jenis Kopi Yang Terjual

```python
df.groupby('kopi')['harga'].sum().to_frame()
```

```python
df['total_omset'] = df.groupby('kopi')['harga'].transform(sum)
df
```
