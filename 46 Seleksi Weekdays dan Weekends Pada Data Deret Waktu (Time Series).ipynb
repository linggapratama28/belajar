```python
import pandas as pd
import numpy as np
```

```python
n_rows = 365
n_cols = 2
cols = ['col1', 'col2']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)

df.index = pd.util.testing.makeDateIndex(n_rows, freq='D') # freq='D' artinya day
df
```

## Seleksi Weekdays dan Weekends

```python
weekdays_df = df[df.index.dayofweek.isin([0, 1, 2, 3, 4])] # 0 berarti senin, 1 berarti selasa, 2 rabu, 3 kamis, 4 jumat
weekdays_df.head(7)
# atribute index.dayofweek hanya bisa diakses kalau memang data frame nya merupakan data frame time series. kalau data frame bukan time series maka tidak bisa di gunakan
# method .isin untuk mengecek apakah dayofweek nya berada pada deret angka seperti yg ada di list yaitu [0, 1, 2, 3, 4]
```

```python
weekends_df = df[df.index.dayofweek.isin([5, 6])] # 5 merepresentasikan sabtu dan 5 merepresentasikan minggu
weekends_df.head(7)
```
